```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
# 创建数据集
data = {
    '进料速率': [100, 150, 200, 250, 300],
    '反应温度': [200, 205, 210, 215, 220],
    '压力': [10, 12, 15, 20, 25],
    '产品产量': [400, 600, 800, 950, 1000],
    '产品质量评分': [8, 7, 6, 5, 4]
}
df = pd.DataFrame(data)
```

```python
# 线性回归模型
# 分别对产品产量和产品质量评分建模
X = df[['进料速率']].values
Y_production = df['产品产量'].values
Y_quality = df['产品质量评分'].values
```

```python
# 产品产量模型
model_production = LinearRegression()
model_production.fit(X, Y_production)
```

```python
# 产品质量评分模型
model_quality = LinearRegression()
model_quality.fit(X, Y_quality)
```

```python
# 绘制图表
plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
plt.scatter(X, Y_production, color='blue', label='实际产量')
plt.plot(X, model_production.predict(X), color='red', label='预测产量')
plt.xlabel('进料速率')
plt.ylabel('产品产量')
plt.title('进料速率与产品产量的关系')
plt.legend()

plt.subplot(1, 2, 2)
plt.scatter(X, Y_quality, color='green', label='实际质量评分')
plt.plot(X, model_quality.predict(X), color='orange', label='预测质量评分')
plt.xlabel('进料速率')
plt.ylabel('产品质量评分')
plt.title('进料速率与产品质量评分的关系')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
# 输出模型系数
model_production_coef = model_production.coef_[0]
model_quality_coef = model_quality.coef_[0]

model_production_coef, model_quality_coef
```
